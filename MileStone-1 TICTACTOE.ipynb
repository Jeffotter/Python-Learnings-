```python
def gameon_choice():
    choice='wrong'
    while choice not in ['Y','N']:
        choice=input("Keep play again? (Y or N): ")
        if choice not in ['Y','N']:
            print("Sorry, I dont know what you are talking about")
        
    if( choice == "Y"):
        return True
    else:
        return False
```
